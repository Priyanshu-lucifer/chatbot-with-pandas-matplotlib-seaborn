```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "User_Input": ["hi", "how are you", "what is your name", "bye", "thanks", "help", "kem palty"],
    "Bot_Response": [
        "Hello! How can I assist you?",
        "I'm a bot, but I'm doing great!",
        "I'm a simple chatbot created using Python!",
        "Goodbye! Aavjoo!",
        "You're welcome!",
        "I can help you with basic questions.",
        "Hu majamaa ....tame kem cho...?"
    ]
}

chatbot_df = pd.DataFrame(data)
print("ðŸ“‹ Sample Chatbot Dataset:\n")
display(chatbot_df)

def chatbot(user_input):
    user_input = user_input.lower()
    if user_input in chatbot_df['User_Input'].values:
        return chatbot_df[chatbot_df['User_Input'] == user_input]['Bot_Response'].values[0]
    else:
        return "I'm sorry, I don't understand that."

print("\nðŸ¤– Welcome to the Chatbot! Type 'exit' to quit.")
while True:
    user_input = input("You: ")
    if user_input.lower() == 'exit':
        print("Bot: Goodbye!")
        break
    response = chatbot(user_input)
    print("Bot:", response)

chatbot_df['Bot_Response_Length'] = chatbot_df['Bot_Response'].apply(len)

plt.figure(figsize=(10, 5))
sns.barplot(x='User_Input', y='Bot_Response_Length', data=chatbot_df, palette='coolwarm')
plt.title('Length of Bot Responses by User Input')
plt.xlabel('User Input')
plt.ylabel('Response Length')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
